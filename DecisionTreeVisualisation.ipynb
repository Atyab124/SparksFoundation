```python
#ATYAB TOSIF
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split
```

## Data preprocessing

```python
#We look at the data and make any reqired changes so as to get optimal predictions when using the classifier
```

```python
#Reading the CSV file
data=pd.read_csv('Iris.csv')
```

```python
data.head()
```

```python
data.dtypes
```

```python
data.columns
```

```python
#Separating the features from the target values and applying train_test_split
features=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
target=['Species']
X=data[features]
y=data[target]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=0)
```

```python
data.Species.unique()
```

## Creating a classifier

```python
#Here we create a decision tree classifier and train in on the train data
classifier=DecisionTreeClassifier()
classifier.fit(X_train,y_train)
```

## Visualising Decision Tree

```python
#Using these lines of code we can see exactly how the decision tree splits the data all the way up to the leaf nodes
#We can also see the entropy function (gini) in each level which determines the purity of the node, the lower the better
fig = plt.figure(figsize=(50,30))
a=tree.plot_tree(classifier,feature_names=features,class_names=data.Species.unique(),filled=True)
```

## Making Predictions

```python
#Here we are using a value from the test data to demonstrate the models ability to make a prediction
classifier.predict(X_test.iloc[[1]])
```
