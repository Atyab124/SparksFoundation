```python
#ATYAB TOSIF
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
data=pd.read_csv('Iris.csv')
```

```python
data.head()
```

```python
data.dtypes
```

```python
data.columns
```

```python
features=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
target=['Species']
X=data[features]
y=data[target]
classifier=DecisionTreeClassifier()
```

```python
data.Species.unique()
```

```python
classifier.fit(X,y)
```

### Visualising

```python
fig = plt.figure(figsize=(50,30))
a=tree.plot_tree(classifier,feature_names=features,class_names=data.Species.unique(),filled=True)
```
